```python
from geopy.geocoders import Nominatim

#Nominatim 객체 생성
geo_local = Nominatim(user_agent = 'South Korea', timeout=None)

#위도/경도 반환 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0,0]
```

```python
geocoding("서울 영등포구 노들로 237")
```

```python
acc = pd.DataFrame(acc['장소설명'])
acc
```

```python
result_df = pd.DataFrame(columns=['Address','Latitude','Longitude'])

for index, row in acc.iterrows():
    address = row['장소설명']
    coordinates = geocoding(address) # 위도/경도 변환
    lat, lng = coordinates
    
    result_df = pd.concat([result_df, pd.DataFrame({'Address':[address], 'Latitude':[lat], 'Longitude':[lng]})], ignore_index=True)
```

```python
result_df.to_csv('자전거 사고다발 구역 위경도_fin.csv', encoding='cp949')
```
